```python
# views 컬럼 - 내림차순 정렬 - 상위 10개 불러오기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")
df = df.sort_values('views', ascending=False)
df.head(10)
```

```python
#10개 중에서 10번째 (최소값) 값 대체
df.iloc[:10,-1] = min_value

#[:10 ,-1] 0행부터 9행까지 마지막 열을 민벨류 값으로 하라는 것
```

## 문제2
- 주어진 데이터셋(members.csv)의 앞에서부터 순서대로 80% 데이터만 활용해 'f1'컬럼 결측치를 중앙값으로 채우기 전 후의 표준편차를 구하고, 두 표준편차 차이 계산하기
(단, 표본표준편차 기준, 두 표준편차 차이는 절대값으로 계산)

풀이 순서
- 앞에서 부터 80% 데이터 뽑아오기
- f1 컬럼 표준편차 구하기
- f1 컬럼 결측치, 중앙값 확인
- f1 컬럼 결측값 중앙값으로 채우기
- f1 컬럼 표준편차 구하기
- 두 표준편차 차이 계산

```python
# 앞에서 부터 80% 데이터 뽑아오기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")

line = int(len(df) * 0.8)
line
df = df.iloc[:line] # 0~79
df
```

```python
## 결측치 채우기 전 f1컬럼 표준편자
std1 = df['f1'].std()
std1
```

```python
## 결측치 확인
df.isnull().sum()
```

```python
## 중앙값 확인
med=df['f1'].median()
med
```

```python
## 중앙값으로 채우기
df['f1'] = df['f1'].fillna(med)
```

```python
## 결측치 확인
df.isnull().sum()
```

```python
## 결측치를 채운 후 표준편차 구하기
std2 = df['f1'].std()
std2
```

```python
# 두 표준편차 차이 계산
print(abs(std1-std2))
```

```python
#내가 만든 한방풀이
df = df.head(80)
print(df.shape)
cond1 = df['f1'].std()
print(cond1)

med = df['f1'].median()
print(df['f1'].isnull().sum().sum())
df['f1'] = df['f1'].fillna(med)
print(df['f1'].isnull().sum().sum())
cond2 = df['f1'].std()
print(cond2)

print(abs(cond1 - cond2))
```

## 문제3
- 주어진 데이터셋(members.csv)의 'age'컬럼의 이상치를 모두 더하시오!
- 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

풀이 순서
- 표준편차, 평균값 구하기
- 이상치 최저, 최고 기준 구하기
- 이상치를 벗어나는 값(조건) 찾아 더하기

```python
# 표준편차, 평균값 구하기
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")

std = df['age'].std() * 1.5
mean = df['age'].mean()

print("표준편차 * 1.5 : ", std)
print("평균 : ", mean)
```

```python
# 이상치 최저, 최고 기준 구하기
lower = mean - std
upper = mean + std

print("이상치 최저 기준 : ", lower)
print("이상치 최고 기준 : ", upper)
```

```python
# 이상치를 벗어나는 값(조건) 찾기

cond1 = df['age'] < lower
cond2 = df['age'] > upper

df[cond1]
```

```python
df[cond2]
```

```python
# 이상치 age합
print(df[cond1|cond2]['age'].sum())
```
