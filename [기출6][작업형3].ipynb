### 1-2. 감기약의 예상 부작용 비율과 항암약의 부작용 관찰값이 통계적으로 유의미하게 차이가 있는지 확인하려 한다. 카이 제곱 검정을 사용하여 검정 통계량을 구하시오.

이 경우는 관찰된 부작용 빈도(항암약)와 기대 빈도(감기약의 예상 부작용 비율에서 환자 수 기반으로 환산) 를 비교하는 것이므로, 일표본 카이제곱 검정(χ² 적합도 검정, goodness-of-fit test)에 해당합니다. 즉, 하나의 분포(예상 비율)에 대해 실제 관찰값이 얼마나 잘 들어맞는지 검정하는 방법입니다.

```python
ob = df['항암약'].value_counts().sort_index().to_list() #이건관찰빈도(observatory)   #인덱스 순으로 하고 , 카이제곱 검정을 하기위해 list형태로 만들어야함!!!   = 이게 오른쪽거랑 같음 = ob=[4,3,2,11]
ex = [0.1 * 20, 0.05 * 20, 0.15 * 20, 0.7 * 20] #이건 기대빈도수(expectaion)   #감기약부작용 발생비율(내가 발생한 실제 비율이 아니고, 우리가 정한 걸로하는것) * 관찰값의전체수(20)

from scipy import stats
stats.chisquare(ob, ex)


#ex의 0.1이런건 우리가 구하려는 기대 비

```

### Q2. 다중 선형 회귀 문제

```python
#다중 선형회귀는 종속변수가 여러가지 독립변수에 얼마나 영향을 받는지 보는것임
# 그래서 밑에 문제는 종속변수인 온도가 태양에너지, 바람세기, 오존농도의 1단위 변화에 따라 어떻게 변하는지를 보는것임!
# 물류로 보면 지급율이, 배차횟수, 장거리근거리여부, 운송시간 등에 따라 지급율이 어떻게 변하는지 상관관계를 확인할 수 있음
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p4/6_3/data6-3-2.csv")
df.head(2)
```

1. 다중 선형 회귀 모델을 구축하고, 독립변수 o3의 회귀계수를 구하시오.
- 독립변수: solar(태양 에너지), wind(바람의 세기), o3(오존 농도)
- 종속변수: temperature(온도)


```python
from statsmodels.formula.api import ols
formula = "temperature ~ solar + wind + o3"
model = ols(formula, data=df).fit()
print(model.summary())
# 0.0749
model.params['o3']
```

2. 데이터에서 'solar'와 'o3' 값을 고정한 상태에서, 'wind'의 세기가 증가함에 따라 'temperature'가 감소하는지를 검증하기 위해 다중 선형 회귀 분석을 수행하고, 이 때'wind'의 회귀 계수에 대한 p-value 값을 구하시오. (유의수준: 0.05)

```python
# 이거 활용해서 기업 규모여부와 재계약의 상관관계등등 확인하면 좋을듯
```

```python
0.78
model.pvalues['wind']
```

3. solar:100, wind:5, o3:30일 때 예측값과 그에 대한 95% 신뢰구간을 구하시오.

```python
new_data = pd.DataFrame({
    'solar':[100],
    'wind':[5],
    'o3':[30]
})
pred = model.get_prediction(new_data)
print(pred.summary_frame(alpha=0.05))
```
