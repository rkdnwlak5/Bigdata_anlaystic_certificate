```python

import pandas as pd

df = pd.read_csv("data/Titanic.csv")


#1번 독립성 검정 / 카이제곱 통계량
from scipy.stats import chi2_contingency
ct = pd.crosstab(df['Gender'], df['Survived'])
chi2, p ,dof, expected = chi2_contingency(ct)
print(chi2_contingency(ct))
print(round(chi2,3))

#2번 로지스틱회귀 / 변수의 계수값
from statsmodels.formula.api import logit
model = logit('Survived ~ Gender + SibSp + Parch + Fare', data = df).fit()
print(model.summary())
print(model.params['Parch'])


#3번 오즈비
import numpy as np
odds = model.params['SibSp']
print(np.exp(odds))
print(round(np.exp(odds),3))
```
