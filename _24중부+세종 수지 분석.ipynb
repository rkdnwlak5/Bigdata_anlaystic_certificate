(1)막대그래프

지점별로 금액 합계를 시각화



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 폰트 설정
plt.rc('font', family='AppleGothic')
plt.rcParams['axes.unicode_minus'] = False

def plot_sales_by_branch(df):
    plt.figure(figsize=(12, 6))
    sales_data = df[df['비용성격'] == '매출액'].groupby('지점명')['금액'].sum()
    sns.barplot(x=sales_data.index, y=sales_data.values)
    plt.title('지점별 총 매출액 비교', fontsize=15)
    plt.ylabel('금액 (원)', fontsize=12)
    plt.xlabel('지점명', fontsize=12)
    for i, v in enumerate(sales_data.values):
        plt.text(i, v, f'{v:,.0f}원', ha='center', va='bottom')
    plt.tight_layout()
    plt.show()

# 함수 실행
plot_sales_by_branch(df)
```
