```python
df[['neighbourhood_group', 'room_type']].groupby(['neighbourhood_group','room_type']).size().unstack()
```

```python
(

df.groupby(['neighbourhood_group', 'room_type'])
  .size()
  .unstack()
  .apply(lambda x: x / x.sum(), axis=1)

)

#여기서 lambda x:는 즉석에서 정의한 함수로 apply함수는 행이나 열을 하나씩 꺼내서 그걸 함수에 넣어준다
#즉 여기서 꺼낸 그 행이 바로 x 다 다시말해 x는 자동으로 행이된다 axis=1이기떄문에, 0이면 열이됨

# axis=1 → 행 기준 → 각 행(row) 하나씩 꺼냄

# x는 그 행 하나임 (Series 형태로 들어옴)

# x / x.sum() → 그 행 전체를 그 행의 합으로 나눔 → 비율 계


#이게 데이터 마님 답인데 맨밑 코드가 어려워서 위에 있는 코드 즉 람다 써서 하기
# Ans = df[['neighbourhood_group','room_type']].groupby(['neighbourhood_group','room_type']).size().unstack()
# Ans.loc[:,:] = (Ans.values /Ans.sum(axis=1).values.reshape(-1,1))
```

# pattern: 바꾸고자 하는 문자(패턴) (정규표현식 가능)

# replacement: 바꾼 후 대체할 문자열

# str_replace()는 처음 등장하는 패턴 하나만 바꾸고, str_replace_all()은 모든 패턴을 바꿉니다.

## 필요하다면 str_replace_all()로도 변경 가능

```python
# Map함수 사용
dic = {
    'Unknown'        : 'N',
    'Less than $40K' : 'a',
    '$40K - $60K'    : 'b',
    '$60K - $80K'    : 'c',
    '$80K - $120K'   : 'd',
    '$120K +'        : 'e'
}

df['newIncome'] = df.Income_Category.map(lambda x: dic[x])

df['newIncome']
```

Q.58 Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
Unknown : N
Less than $40K : a
$40K - $60K : b
$60K - $80K : c
$80K - $120K : d
$120K +’ : e

```python
#58 apply 함수 사용

def changeCategory(x) :
  if x == "Unknown" :
      return  "N"
  elif x == "Less than $40K" :
      return 'a'
  elif x == "$40K - $60K" :
      return 'b'
  elif x == "$60K - $80K" :
      return 'c'
  elif x == "$80K - $120K" :
      return 'd'
  elif x == "$120K +" :
      return "e"

df["newIncome"] = df["Income_Category"].apply(changeCategory)
df["newIncome"]
```

Q.62 Marital_Status 컬럼값이 Married 이고 Card_Category 컬럼의 값이 Platinum인 경우 1 그외의 경우에는 모두 0으로 하는 newState컬럼을 정의하라. newState의 각 값들의 빈도수를 출력하라

76.년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(":")[0]

    if hour =='24':
        hour ='00:00:00'

        FinalDate = pd.to_datetime(date +" "+hour) +datetime.timedelta(days=1)

    else:
        hour = hour +':00:00'
        FinalDate = pd.to_datetime(date +" "+hour)

    return FinalDate

df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)

df



# hour = x.split(':')[1]
 # date = x.split(":")[0]
 #여기 []안에 숫자 의미는
 #x.split(':')를 나누면 ['2024-01-05', '14'] 이렇게 나누어짐 여기서 0번쨰는 첫번쨰, 1번쨰는 두번쨰니까
 #즉 몇번째 요소를 가져올지를 의미하는 것임
```
