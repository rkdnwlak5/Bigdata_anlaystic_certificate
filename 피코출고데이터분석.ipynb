```python
파일명들 = ['df_24년1월', 'df_24년2월', 'df_24년3월', 'df_24년4월', 'df_24년5월', 'df_24년6월', 'df_24년7월', 'df_24년8월', 'df_24년9월', 'df_24년10월', 'df_24년11월', 'df_24년12월' ]  # 문자열 리스트
df_list = [eval(name) for name in 파일명들]  # 또는 safer: [dfs[name] for name in 파일명들]
df_24all = pd.concat(df_list, axis=0, ignore_index=True)  # ignore_index로 인덱스 재정렬

파일명25년 = ['df_25년1월', 'df_25년2월', 'df_25년3월', 'df_25년4월', 'df_25년5월', 'df_25년6월']
df_list2 = [eval(name) for name in 파일명들]  # 또는 safer: [dfs[name] for name in 파일명들]
df_25all = pd.concat(df_list2, axis=0, ignore_index=True)  # ignore_index로 인덱스 재정렬
```

```python
#24년데이터셋 전처리
import pandas as pd

# 1. 쉼표 제거 후 숫자형으로 변환할 컬럼
comma_cols = ['예정수량', '등록수량', '확정수량', '공급금액', '확정BOX', '확정EA', '확정중량']
for col in comma_cols:
    if col in df_24all.columns:
        df_24all[col] = df_24all[col].astype(str).str.replace(',', '', regex=False)
        df_24all[col] = pd.to_numeric(df_24all[col], errors='coerce')

# 2. 날짜형 컬럼 변환
date_cols = ['출고일자', '지시일시', '검수일시', '확정일시', '등록일시', '전표일자']
for col in date_cols:
    if col in df_24all.columns:
        df_24all[col] = pd.to_datetime(df_24all[col], errors='coerce')

# 3. 숫자형 컬럼 결측값 0으로 채우기
num_cols = df_24all.select_dtypes(include=['int', 'float']).columns
df_24all[num_cols] = df_24all[num_cols].fillna(0)

# 4. 문자열(object)형 결측값 '미기입'으로 채우기
obj_cols = df_24all.select_dtypes(include='object').columns
df_24all[obj_cols] = df_24all[obj_cols].fillna('미기입')


# 5. 날짜형 결측값을 '미기입'으로 채우기 위해 object 형으로 변환 후 채움
date_cols = ['전표일자', '등록일시', '지시일시', '검수일시', '확정일시', '출고일자']
for col in date_cols:
    if col in df_24all.columns:
        df_24all[col] = df_24all[col].astype(str)
        df_24all[col] = df_24all[col].replace('NaT', '미기입')  # NaT 문자열화된 것 처리
        df_24all[col] = df_24all[col].replace('nan', '미기입')  # NaN 처리 (float -> str 이후)
        df_24all[col] = df_24all[col].fillna('미기입')          # 혹시 남아있을 결측값도

# 결과 확인
print(df_24all[date_cols].isnull().sum())


# 결과 확인
print(df_24all.dtypes)
print(df_24all.isnull().sum())



#미기입 및 NaN 문자 열 등은 자동으로 NaN 처리해서 sum등 계산 가능하게 선작업
df_24all['확정수량'] = pd.to_numeric(df_24all['확정수량'], errors='coerce')





# 파일다운받으려면
# df_24all.to_excel("전처리_완료.xlsx", index=False)
```
