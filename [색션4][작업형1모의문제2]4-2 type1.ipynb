## 문제5
- 데이터에서 IQR을 활용해 views 컬럼의 이상치를 찾고, 이상치 데이터의 수를 구하시오

````python
# your code
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")

#IQR 구하기 위해서는 3사분위 75% 지점과, 1사분위 25% 지점를 구해야함

q3 = df['views'].quantile(.75)
q1 = df['views'].quantile(.25)

IQR = q3 - q1
# print(IQR)

#이상치는 1사분위 -1.5*IQR 와 3사분위 +1.5*IQR

line1 = q1 - 1.5 * IQR
line2 = q3 + 1.5 * IQR

print(sum(df['views'] < line1) + sum(df['views'] > line2))


````

## 문제6
- 'age' 컬럼의 이상치(소수점 나이와 음수나이, 0포함)를 제거하고
- 제거 전 후의 views 컬럼 표준편차를 더하시오 (최종 결과 값은 소수 둘째자리까지 출력, 셋째자리에서 반올림)

````python
# your code
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")
df

#제거 전 표준편차
r1 = df['views'].std()


# 음수나이, 0포함 제거
# 앞에 ~ 표시하면 이값들이 아닌것임
cond = df['age'] <= 0
# print(df.shape)
df = df[~cond]
# print(df.shape)

# 소수점 나이 제가
# 반올림해서 원래랑 숫자가 다르면 소수점 나이라고 하자
# != 같지않다로 하면 소수점 나이 찾아짐, 그냥 == 으로하면 정수형만 나와서 소수점이 다 배제 됨
# print(df.shape)
cond = df['age'] == round(df['age'],0)
df = df[cond]
# print(df.shape)

r2 = df['views'].std()

#변경 사항 저장전 반드시 shape 보는 습관 들이기

print(round(r1+r2, 2))
````

## 문제6
- 'age' 컬럼의 이상치(소수점 나이와 음수나이, 0포함)를 제거하고
- 제거 전 후의 views 컬럼 표준편차를 더하시오 (최종 결과 값은 소수 둘째자리까지 출력, 셋째자리에서 반올림)

````python
# 풀이
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/main/p1/members.csv")

r1 = df['views'].std()

cond = df['age'] <= 0
# print(df.shape)
df = df[~cond]
# print(df.shape)

# print(df.shape)
cond = df['age'] == round(df['age'],0)
df = df[cond]
# print(df.shape)

r2 = df['views'].std()

print(round(r1 + r2, 2))
````
