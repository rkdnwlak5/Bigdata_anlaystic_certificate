```python
#로또 만들기
import random

# 중복되지 않은 로또 번호를 담을 리스트
lotto_numbers = []

# 번호를 6개 뽑을 때까지 반복
while len(lotto_numbers) < 6:
    i = random.randint(1, 45)  # 1부터 45까지 랜덤 숫자 뽑기
    if i not in lotto_numbers:  # 중복을 확인하고, 중복되지 않으면 리스트에 추가
        lotto_numbers.append(i)

# 결과 출력
print(sorted(lotto_numbers))  # 오름차순으로 출력


```

```python
#가위바위보를 컴퓨터 끼리하도록 하기

import random
toss = []

for i in range(10000) :
    comA = random.choice(["가위", "바위", "보"])
    comB = random.choice(["가위", "바위", "보"])

    if comA == "가위" :
        if comB == "가위" :
            toss.append("없음")
        elif comB == "바위" :
            toss.append("B")
        elif comB == "보" :
            toss.append("A")

    if comA == "바위" :
        if comB == "바위" :
            toss.append("없음")
        elif comB == "보" :
            toss.append("B")
        elif comB == "가위" :
            toss.append("A")

    if comA == "보" :
        if comB == "보" :
            toss.append("없음")
        elif comB == "가위" :
            toss.append("B")
        elif comB == "바위" :
            toss.append("A")

a_win = toss.count("A")
b_win = toss.count("B")
draw = toss.count("없음")

print("10000번 중 컴퓨터 A의 승리 :", a_win)
print("10000번 중 컴퓨터 B의 승리 :", b_win)
print("10000번 중 비긴 경기 :", draw)
```
