```python
import pandas as pd

data = {
    '상품명': ['크리스마스트리', '산타모자', '선물상자', '장식품', '캔디케인', '산타양말', '눈사람모형'],
    '가격': [50, 10, 20, 15, 5, 12, 30],
    '판매량': [200, 500, 300, 400, 1000, 300, 150],
    '카테고리': ['장식', '의류', '포장', '장식', '간식', '의류', '장식']
}

df = pd.DataFrame(data)
```

```python
# 단일/복수 컬럼 선택
cols = ['상품명','가격']
type(df[cols])
```

```python
# Scaler
#스케일링할때는 cols = df['뭐뭐'] 이런식으로 안넣고 하니까 뒤에 fit_transform할때 df하고 [[]] 괄호 두개 꼭 하는거 잊지 않기!
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['가격_scaled'] = scaler.fit_transform(df[['가격']])
df
```

```python
# Scaler2
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
cols = ['가격', '판매량']
df[cols] = scaler.fit_transform(df[cols])
df
```
