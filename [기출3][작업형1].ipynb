## 문제1
- 결측치 데이터(행)을 제거하고, 앞에서부터 70% 데이터만 활용해, 'f1' 컬럼 1사분위 값을 구하시오
- 데이터: members.csv

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/inf/refs/heads/main/p1/members.csv")
```

```python
#결측치 삭제는 dropna
#몇퍼센트 개수 등 -> len쓰고 소수점 나오면 int
#70%까지 잘라내기 = iloc
print(df.shape)
print(df.isnull().sum())
df = df.dropna()
print(df.shape)

print(int(len(df)*0.7))
df = df.iloc[:int(len(df)*0.7)]
print(df.shape)

print(df['f1'].quantile(.25))


```

## 문제2
- index는 년도임. 2000년 데이터 중 2000년 평균보다 큰 값의 데이터 수
- 데이터: 아래 셀 실행

```python
# 데이터 생성(먼저 실행해 주세요)
import pandas as pd
import random
random.seed(2022)
df = pd.DataFrame()
for i in range(0, 3):
    list_box = []
    for k in range(0, 200):
        ran_num = random.randint(1,200)
        list_box.append(ran_num)
    df[i+1999] = list_box
df = df.T
df
```

```python
#방법1
#컬럼의 평균은 많이 구해보았으나, 행의 평균은 안구해봄 -> loc 또는 iloc 를 사용 해야함
mean = df.loc[2000].mean()
print(sum(df.loc[2000,:] > mean      # 이거는 2000이라는 행에서, 전체 값중 평균보다 큰값을 더하라는 것
# print(len(df.loc[2000] > mean))

```

```python
#방법2
#행과 열을 바꾼 후 푸는 방법
#df.T때문에 실행하면 다시 원래대로 가로로 돌아가기 떄문에 위에 코드(문제) 실행 후 해야만함!
df = df.T
df
m = df[2000].mean()

print(sum(df[2000] > m))
```
