## 시계열 데이터 (Timedelta)

```python
# 데이터 불러오기 (usecols, parse_dates)
# 여기서 데이터 불러올 때 전체를 불러오는게 아닌 특정 컬럼 한개만 불러보는 방법
#usecols를 사용해서 진행
df = pd.read_csv("date.csv", usecols= ['DateTime4'], parse_dates = ['DateTime4'])
df.info()
```

# 데이터 타임은 특정시점의 날짜와 시간을 의미
# 타임 델타는 두 시점의 차이를 의미

```python
# 100일째 되는 날
day = pd.Timedelta(days=99)
df['100일']= df['DateTime4'] + day
df
```

```python
# 100시간 이후
hour = pd.Timedelta(hours=100)
df['100시간']= df['DateTime4'] + hour
df
```

```python
# 시간 +/- (3주, 3일, 3시간, 3분, 3초 이전과 이후)
# timedelta 쓸때 주의 할점은 뒤에 다 s를 붙여야함 days, hours 등등
diff = pd.Timedelta(weeks=3, days=3, hours=3, minutes=3, seconds=3)
df['+diff'] = df['DateTime4'] + diff
df['-diff'] = df['DateTime4'] - diff
df
```

```python
# timedelta 자료형(datetime간의 연산 결과)
diff = df['100일'] - df['100시간']
diff
```

```python
# 기간을 초로 변환 total_seconds()
# 여기서는 분만 계산해서 변환 할 수 있는 함수는 없음, 그래서 분을 물어봐도 그걸 초로 변환해서 계산하면됨
# totalseconds는 timedelta 에서만 사용가능함 -> 초로만 변경된 값
print(diff.dt.total_seconds()) #초
print(diff.dt.total_seconds()/60) #분
print(diff.dt.total_seconds()/60/60) #시간
print(diff.dt.total_seconds()/60/60/24) #일
```

```python
# 일(days), 초(seconds)
print(diff.dt.days)
print(diff.dt.seconds)

#위의 total_seconds와 다른 이유
```

#위의 total_seconds와 다른 이유
Total seconds는 94일 20시간을 모두 초로 변경을 한 값이고
diff.dt.seconds는 여기에 있는 20시간만 초로 변경을 한것임
즉) 94일은 제외하고 시간만 나온것
